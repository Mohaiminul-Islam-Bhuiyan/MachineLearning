```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Dhaka Rent.csv")
```

```python
df.head()
```

```python
plt.scatter(df.area , df.rent)
plt.plot(df.area, df.rent)
plt.xlabel("area")
plt.ylabel("rent")
plt.title('area vs rent in dhaka')
```

```python
# df.info
```

###### Sep data for train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['area']] # x = df.drop ('rent', axis = 1)
y = df[['rent']] # y = df.drop ('area', axis = 1)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = .30, random_state = 1) #train_size = .7
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtrain.to_csv('train_x_csv')
```

```python
from sklearn.linear_model import LinearRegression as lr
```

```python
reg = lr()
```

```python
reg.fit(xtrain, ytrain)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2000]])
```

```python
xtrain.shape
```

```python
reg.score(xtest, ytest)
```

```python
plt.plot(xtest, reg.predict(xtest))
plt.scatter(xtest, ytest, color = "blue")
plt.scatter(xtrain, ytrain, color = "red")
```
