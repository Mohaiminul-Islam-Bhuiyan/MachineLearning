```python
x = df["x"]
y = df["y"]
```

# linear regression using gradient decent

```python
m = 0
c = 0
l = 0.00001

iteration = 10000

n = int(len(x))

for i in range (iteration):
    pred = (m * x) + c
    der_m = (-2/n) * sum(x*(y-pred))
    der_c = ((-2/n) * sum(y - pred))
    
    m = m- (l*der_m)
    c = c- (l*der_c)
    
print (m,c)
```
